```python
import numpy as np
```

```python
# Кількість елементів у векторі (можна змінити за потребою)
n = 1000

# Створення вектора з квадратними коренями
vector = np.sqrt(np.arange(1, n + 1))

# Виведення вектора
print(vector)
```

```python
# Створення двовимірного масиву розміром 3x3 з залишками від ділення на 3
matrix = np.arange(1, 10).reshape(3, 3) % 3

# Виведення матриці
print(matrix)
```

```python
# Розмірність масиву
rows, cols = 9, 9

# Створення масиву розміром 9x9 з випадковими натуральними числами від 1 до 50
matrix = np.random.randint(1, 51, size=(rows, cols))

# Знайдемо відсоток чисел, які діляться на 3 без остачі
divisible_by_3 = (matrix % 3 == 0)
percentage_divisible_by_3 = (np.sum(divisible_by_3) / (rows * cols)) * 100

# Виведення результатів
print("Матриця:")
print(matrix)
print("\nВідсоток чисел, які діляться на 3 без остачі:", percentage_divisible_by_3, "%")
```

```python
# Створення двох масивів розміром 7 з випадковими цілими числами від 1 до 5
array1 = np.random.randint(1, 6, size=7)
array2 = np.random.randint(1, 6, size=7)

# Виведення двох масивів
print("Перший масив:", array1)
print("Другий масив:", array2)

# Виконання поелементного додавання, віднімання та множення
addition_result = array1 + array2
subtraction_result = array1 - array2
multiplication_result = array1 * array2

# Виведення результатів
print("Результат додавання:", addition_result)
print("Результат віднімання:", subtraction_result)
print("Результат множення:", multiplication_result)
```

```python
# Створення двох векторів розміром 4 з довільними дійсними числами
vector1 = np.array([2.0, 3.0, 1.5, 4.2])
vector2 = np.array([1.0, 2.0, 0.5, 3.0])

# Обчислення скалярного добутку
scalar_product = np.dot(vector1, vector2)

# Виведення результату
print("Перший вектор:", vector1)
print("Другий вектор:", vector2)
print("Скалярний добуток:", scalar_product)
```

```python
# Створення двох матриць розміром 2x3 та 3x4 з випадковими цілими числами від 1 до 15
matrix1 = np.random.randint(1, 16, size=(2, 3))
matrix2 = np.random.randint(1, 16, size=(3, 4))

# Виведення двох матриць
print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)

# Перемноження матриць
result_matrix = np.dot(matrix1, matrix2)

# Виведення результату
print("\nРезультат перемноження матриць:")
print(result_matrix)
```

```python
matrix = np.random.randint(1, 16, size=(2,3))

# Виведення початкової матриці
print("Початкова матриця:")
print(matrix)

# Знаходження оберненої матриці
try:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("\nМатриця не має оберненої матриці (матриця сингулярна).")
```

```python
# Створення матриці розміром 3x3 з випадковими цілими числами від 3 до 15
matrix = np.random.randint(3, 16, size=(3, 3))

# Виведення початкової матриці
print("Початкова матриця:")
print(matrix)

# Транспонування матриці
transposed_matrix = np.transpose(matrix)

# Виведення транспонованої матриці
print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
# Створення матриці розміром 3x4 з випадковими цілими числами від 1 до 10
matrix = np.random.randint(1, 11, size=(3, 4))

# Створення вектора розміром 4 з випадковими цілими числами від 1 до 10
vector = np.random.randint(1, 11, size=4)

# Виведення матриці і вектора
print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

# Обчислення добутку матриці на вектор
result = np.dot(matrix, vector)

# Виведення результату
print("\nРезультат добутку матриці на вектор:")
print(result)
```

```python
# Створення матриці розміром 8x8 з випадковими цілими числами від 1 до 10
matrix1 = np.random.randint(1, 11, size=(8, 8))
# Створення матриці 2 розміром 8x8 з випадковими цілими числами від 1 до 10
matrix2 = np.random.randint(1, 11, size=(8, 8))

# Виведення матриці
print("Матриця:")
print(matrix1)

# Знаходження добутку всіх елементів матриці
product_of_elements = np.prod(matrix1)

# Виведення результату
print("\nДобуток всіх елементів матриці:", product_of_elements)
```

```python
# Створення двох матриць розміром 7x7 з випадковими дійсними числами від -1 до 1
matrix1 = np.random.uniform(-1, 1, size=(7, 7))
matrix2 = np.random.uniform(-1, 1, size=(7, 7))

# Виведення першої матриці
print("Перша матриця:")
print(matrix1)

# Виведення другої матриці
print("\nДруга матриця:")
print(matrix2)

# Знаходження різниці елементів матриць
difference_matrix = matrix1 - matrix2

# Виведення результату
print("\nРізниця елементів матриць:")
print(difference_matrix)
```

```python
# Створення матриці розміром 9x9 з випадковими дійсними числами від -10 до 10
matrix = np.random.uniform(-10, 10, size=(9, 9))

# Виведення матриці
print("Матриця:")
print(matrix)

# Знаходження суми елементів на головній діагоналі
diagonal_sum = np.trace(matrix)

# Виведення результату
print("\nСума елементів на головній діагоналі:", diagonal_sum)
```

```python
# Створення матриці розміром 9x9 з випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, size=(9, 9))

# Виведення матриці (не обов'язково)
print("Матриця:")
print(matrix)

# Знаходження вектору-стовпчика з сумами елементів кожного стовпця
column_sums = np.sum(matrix, axis=0)

# Виведення результату
print("\nВектор-стовпчик з сумами елементів кожного стовпця:")
print(column_sums)
```
